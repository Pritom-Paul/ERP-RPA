```python
# seek
import fitz  # PyMuPDF
import re
import sys

def replace_phone_number(input_pdf, output_pdf, old_phone, new_phone):
    """
    Replace phone numbers in a PDF document
    
    Args:
        input_pdf (str): Path to input PDF file
        output_pdf (str): Path to output PDF file
        old_phone (str): Phone number pattern to replace
        new_phone (str): New phone number to write
    """
    # Open the PDF document
    doc = fitz.open(input_pdf)
    
    # Compile regex pattern for the phone number
    # This pattern will match various phone number formats
    phone_pattern = re.compile(re.escape(old_phone), re.IGNORECASE)
    
    # Iterate through each page
    for page_num in range(len(doc)):
        page = doc[page_num]
        
        # Search for the phone number text
        text_instances = page.search_for(old_phone)
        
        # Remove each found instance
        for inst in text_instances:
            # Add a white rectangle to cover the old phone number
            page.add_redact_annot(inst, fill=(1, 1, 1))  # White fill
            
        # Apply the redactions (remove the text)
        page.apply_redactions()
        
        # Now add the new phone number at the first found location
        if text_instances:
            # Use the position of the first found instance
            rect = text_instances[0]
            
            # Insert the new text at the same position
            page.insert_text(
                (rect.x0, rect.y0+7),  # Position
                new_phone,           # New text
                fontsize=8,         # Adjust font size as needed
                color=(0, 0, 0)      # Black text
            )
    
    # Save the document
    doc.save(output_pdf)
    doc.close()
    print(f"Successfully processed PDF. Saved as: {output_pdf}")

if __name__ == "__main__":
    # Example usage
    input_file = r"C:\Users\Altersense\Desktop\ERP-RPA\Sample\Other Samples\LIDL Polo_55th Round_Manami_JF-05B41688_Light Blue_DE_PO.pdf"
    output_file = r"C:\Users\Altersense\Desktop\ERP-RPA\Sample\Other Samples\LIDL Polo_55th Round_Manami_JF-05B41688_Light Blue_DE_PO_edited.pdf"
    old_phone = "+49 4181 / 9422 - 293"
    new_phone = "new_phone_number"
    
    # Replace with your actual file paths and phone numbers
    replace_phone_number(input_file, output_file, old_phone, new_phone)
```
